## Accuracy

```python
#simply pass ur training data to .score method
knn2.score(X_test, y_test)
```

```python
knn3 = knc(n_neighbors = 3)
knn3.fit(X_train, y_train)
knn3.score(X_test, y_test)
```

```python
knn4 = knc(n_neighbors = 12)
knn4.fit(X_train, y_train)
knn4.score(X_test, y_test)
```

## findind the best fit model

```python
# Hyperparameter Tuning 
# We have to find, number of neighbors, which will give maximum accuracy

accuracies=[]

for i in range(1, 31):
    knn= KNeighborsClassifier(n_neighbors=i) # 1 2 3
    knn.fit(X_train, y_train)
    acc=knn.score(X_test, y_test)
    accuracies.append(acc)
```

```python
accuracies
```

```python
import matplotlib.pyplot as plt
plt.scatter(range(1,31), accuracies)
```
